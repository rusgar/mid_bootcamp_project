```python
def total_goles(equipo):
    goles_local = df[(df['team_name_home'] == f'{equipo}')]['team_away_score'].sum()
    goles_vist = df[(df['team_name_away'] == f'{equipo}')]['team_home_score'].sum()
    return int(goles_local + goles_vist)

goles_encajados = [total_goles(equipo) for equipo in participantes]
goles_encajados 
```

```python
def tiros(equipo):
    tiros_local = df[(df['team_name_home'] == f'{equipo}')]['total_shots_home'].sum()
    tiros_vist = df[(df['team_name_away'] == f'{equipo}')]['total_shots_away'].sum()
    return int(tiros_local + tiros_vist)
tiros_total = [tiros(equipo) for equipo in participantes]
tiros_total
```

```python
def penaltis(equipo):
    penaltis_local = int(df[(df['team_name_home'] == f'{equipo}') & (df['pens_home_score'] != 'False')]['pens_home_score'].sum())
    penaltis_vist = int(df[(df['team_name_away'] == f'{equipo}') & (df['pens_home_score'] != 'False')]['pens_away_score'].sum())
    return int(penaltis_local + penaltis_vist)
penaltis_total = [penaltis(equipo) for equipo in participantes]
penaltis_total
```

```python
def posesion(equipo):
    posesion_local = df[(df['team_name_home'] == f'{equipo}')]['possession_home'].sum()
    posesion_vist = df[(df['team_name_away'] == f'{equipo}')]['possession_away'].sum()
    return int(posesion_local + posesion_vist)
posesion_total = [posesion(equipo) for equipo in participantes]
posesion_total
```

```python
def posesion(equipo):
    pos_local = df[df['team_name_home'] == f'{equipo}']['possession_home'].sum()
    pos_vist = df[df['team_name_away'] == f'{equipo}']['possession_away'].sum()
    partidos =  df[(df['team_name_away'] == f'{equipo}')|(df['team_name_home'] == f'{equipo}')]['stage'].count()
    posesion_equipo = (pos_local+pos_vist)/partidos
    return int(posesion_equipo)

 
posesion_por_partido =  [posesion(equipo) for equipo in participantes]
posesion_por_partido 


```

```python
def entradas(equipo):
    ent_local = df[df['team_name_home'] == f'{equipo}']['duels_won_home'].sum()
    ent_vist = df[df['team_name_away'] == f'{equipo}']['duels_won_away'].sum()
    partidos =  df[(df['team_name_away'] == f'{equipo}')|(df['team_name_home'] == f'{equipo}')]['stage'].count()
    lances_ganados = (ent_local+ent_vist)/partidos
    return int(lances_ganados)

 
lances_por_partido =  [entradas(equipo) for equipo in participantes]
lances_por_partido
```
